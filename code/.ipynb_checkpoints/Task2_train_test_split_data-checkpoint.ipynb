```python
import os
from glob import glob
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
from moviepy.editor import AudioFileClip
from tqdm import tqdm

if __name__ == '__main__':
    path = os.path.join(os.getcwd(),'kickstarter_data')
    fail_files = glob(os.path.join(path, 'failed', '*'))
    sucess_files = glob(os.path.join(path, 'successful', '*'))

    label_0 = [0] * len(fail_files)
    label_1 = [1] * len(sucess_files)

    data = fail_files + sucess_files
    label = label_0 + label_1

    print(len(data))

    new_data = []
    new_label = []
    for i in range(len(data)):
        video_path = glob(os.path.join(data[i], '*.mp4'))
        if len(video_path) == 0:
            continue
        else:
            video_path = video_path[0]
        root = video_path.split('.mp4')[0]
        wav_name = root + '.wav'
        if not os.path.exists(wav_name):
            try:
                my_audio_clip = AudioFileClip(video_path)
                my_audio_clip.write_audiofile(wav_name)
            except:
                continue

        new_data.append(data[i])
        new_label.append(label[i])

    print('data file nums: ', len(new_data))
    print('label nums: ', len(new_label))

    assert len(new_data) == len(new_label), 'data file nums not equal label nums'

    train_x, test_x, train_y, test_y = train_test_split(new_data, new_label, test_size=0.2, shuffle=True)

    #train_x = train_x[ : len(train_x) // 1000]
    #train_y = train_y[: len(train_y) // 1000]
    cols = ['file_path', 'label']
    dataFrame = np.array([train_x, train_y]).T
    dataFrame = pd.DataFrame(data=dataFrame, columns=cols)
    dataFrame.to_csv('train.csv')

    #test_x = test_x[: len(test_x) // 3]
    #test_y = test_y[: len(test_y) // 3]
    #test_x = test_x[: 4]
    #test_y = test_y[: 4]
    dataFrame = np.array([test_x, test_y]).T
    dataFrame = pd.DataFrame(data=dataFrame, columns=cols)
    dataFrame.to_csv('test.csv')



```
